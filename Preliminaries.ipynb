```julia
import Pkg;
Pkg.add("PyPlot")

using PyPlot
```

```julia
using DelimitedFiles
using DataFrames
```

```julia
N = 100;
x = 0:(2*pi/N):2*pi;
y = exp.(-6 * sin.(x));

plot(x,y,"g",linewidth=4)

xlabel("x", fontsize=14, fontweight="bold")
ylabel("y", fontsize=14, fontweight="bold")
grid(true)
title("x vs y", fontsize=18, fontweight="bold");

```
